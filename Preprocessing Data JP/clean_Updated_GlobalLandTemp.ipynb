```python
# Import our dependencies
import pandas as pd
import numpy as np
```

```python
#import csv file

file_path = "Resources/Updated_GlobalLandTemperaturesByState.csv"
global_temp_states = pd.read_csv(file_path)
global_temp_states.head()
```

```python
#list_of_years --> the variable that will hold all the unique years in global_temp_states['Year'] from lowest to highest
    #Extract from global_temp_states and then store into an array/list
    
list_of_years = global_temp_states.Year.unique()
```

```python
# list_of_temp_avgs --> the variable that will hold the array of average temperatures for each year in ascending order (avg temps starting from year 1743 to 2013)
list_of_temp_avgs=[]
for i in range(len(list_of_years)):
    saved_avg = global_temp_states.loc[global_temp_states['Year'] == list_of_years[i], 'AverageTemperature'].mean()
    list_of_temp_avgs.append(saved_avg)
    
list_of_temp_avgs
```

```python
# list_of_unc_avgs --> variable that will hold the array of avg. temp uncertainity in ascending order 
list_of_unc_avgs=[]

for i in range(len(list_of_years)):
    saved_unc_avg = global_temp_states.loc[global_temp_states['Year'] == list_of_years[i], 'AverageTemperatureUncertainty'].mean()
    list_of_unc_avgs.append(saved_unc_avg)
list_of_unc_avgs
```

```python
# Create new dataframe - a new house for our new information
global_temp_usa = pd.DataFrame(columns = ['Year', 'Country', 'Avg Temperature', 'Avg Temperature Uncertainty'])

```

```python
for i in range(len(list_of_years)):
    global_temp_usa = global_temp_usa.append({'Year' : list_of_years[i],
                                              'Country' : 'United States',
                                              'Avg Temperature' : list_of_temp_avgs[i],
                                              'Avg Temperature Uncertainty' : list_of_unc_avgs[i]},
                                             ignore_index=True)
    
global_temp_usa.head(10)
global_temp_usa.shape
```

```python
global_temp_usa = global_temp_usa.drop_duplicates(subset=['Year'])
global_temp_usa.shape
```

```python
# Saving data
file_path = "global_temp_avg_1743_2013.csv"
global_temp_usa.to_csv(file_path, index=False)
```
